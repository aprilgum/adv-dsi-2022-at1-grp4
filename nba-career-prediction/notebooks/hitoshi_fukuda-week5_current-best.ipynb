## 7. Hyperparameter tuning with Hyperopt

```python
#Import Trials, STATUS_OK, tpe, hp, fmin from hyperopt package
!pip install hyperopt
from hyperopt import Trials, STATUS_OK, tpe, hp, fmin
```

```python
#Define the search space for XGBoost hyperparameters
space = {
    'max_depth' : hp.choice('max_depth', range(1, 20, 1)),
    'learning_rate' : hp.quniform('learning_rate', 0.01, 0.5, 0.05),
    'min_child_weight' : hp.quniform('min_child_weight', 1, 10, 1),
    'subsample' : hp.quniform('subsample', 0.1, 1, 0.05),
    'colsample_bytree' : hp.quniform('colsample_bytree', 0.1, 1.0, 0.05)
}
```

```python
#Difine a function that is called 'objective' with the following logics
def objective(space):
    from sklearn.model_selection import cross_val_score
    
    xgboost = xgb.XGBClassifier(
        max_depth = int(space['max_depth']),
        learning_rate = space['learning_rate'],
        min_child_weight = space['min_child_weight'],
        subsample = space['subsample'],
        colsample_bytree = space['colsample_bytree']
    )
    
    acc = cross_val_score(xgboost, X_train, y_train, cv=10, scoring="accuracy").mean()

    return{'loss': 1-acc, 'status': STATUS_OK }
```

```python
#Launch Hyperopt search and save the result in a variable that is called 'best'
best = fmin(
    fn=objective,   
    space=space,       
    algo=tpe.suggest,       
    max_evals=5
)
```

```python
#Print the best set of hyperparameters
print("Best: ", best)
```

```python
#Instantiate a XGBClassifier with best set of hyperparameters
xgbclsfr2 = xgb.XGBClassifier(
    max_depth = best['max_depth'],
    learning_rate = best['learning_rate'],
    min_child_weight = best['min_child_weight'],
    subsample = best['subsample'],
    colsample_bytree = best['colsample_bytree']
)
```

```python
#Fit the model with the prepared data
xgbclsfr2.fit(X_train, y_train)
```

```python
#Display the accuracy and f1 scores of this baseline model on the training and validation sets
print_class_perf(y_preds=xgbclsfr2.predict(X_train), y_actuals=y_train, set_name='Training', average='weighted')
print_class_perf(y_preds=xgbclsfr2.predict(X_val), y_actuals=y_val, set_name='Validation', average='weighted')
```

## 10. Partial Dependence Plot

```python
#Import plot_partial_dependence from sklearn.inspection
from sklearn.inspection import plot_partial_dependence
```

```python
"""
Define a function called plot_pdp
- input parameters: trained model (`model`), features (`X`), name of feature to be analysed (`feature_name`), name of features (`feature_cols`), values of target (`target_classes`)
- logics: find the index of the feature to be analysed and print the partial dependence plot for each value of the target class
"""
def plot_pdp(model, X, feature_name, feature_cols, target_classes):
    feature_index = feature_cols.get_loc(feature_name)
    
    print(f"PDP for `{feature_cols[feature_index]}` ({feature_index}) with {target_classes}")
    
    for target_class in target_classes:
        plot_partial_dependence(model, X, features=[feature_index], target=target_class)
```

```python
#Create a list called target_classes contating the list of all values from the target variable
target_classes=[0, 1, 2, 3]
```

```python
#Display the partial dependence plot for "FTM" feature on the training set with plot_pdp function
plot_pdp(model=xgbclsfr2, X=X_train, feature_name='GP', feature_cols=df_train_tmp.columns, target_classes=target_classes)
```

```python
space2 = {
    'max_depth' : hp.choice('max_depth', range(5, 20, 1)),
    'learning_rate' : hp.quniform('learning_rate', 0.01, 0.5, 0.05),
    'min_child_weight' : hp.quniform('min_child_weight', 1, 10, 1),
    'subsample' : hp.quniform('subsample', 0.1, 1, 0.05),
    'colsample_bytree' : hp.quniform('colsample_bytree', 0.1, 1.0, 0.05)
}
```

```python
def objective(space2):
    from sklearn.model_selection import cross_val_score
    
    xgbclsfr_re = xgb.XGBClassifier(
        max_depth = int(space2['max_depth']),
        learning_rate = space2['learning_rate'],
        min_child_weight = space2['min_child_weight'],
        subsample = space2['subsample'],
        colsample_bytree = space2['colsample_bytree']
    )
    
    acc = cross_val_score(xgbclsfr_re, X_train2, y_train2, cv=10, scoring="accuracy").mean()

    return{'loss': 1-acc, 'status': STATUS_OK }
```

```python
#Launch Hyperopt search and save the result in a variable that is called 'best'
best2 = fmin(
    fn=objective,   
    space=space2,       
    algo=tpe.suggest,       
    max_evals=5
)
```
