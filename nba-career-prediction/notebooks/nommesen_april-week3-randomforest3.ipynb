# Random Forest Classifier with Balanced Weights

## No standardisation

Using  Random Forest Classifier to handle imbalanced data and to feature select

Pre-requisite: nommesen_april-week2-prepare_data.ipynb

```python
import pandas as pd
import numpy as np
import math

# SKLearn
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score

import matplotlib.pyplot as plt
import seaborn as sns

import joblib
```

## Loading the training data set

```python
alltrain = np.load('../data/processed/alltrain.npy')

print("Dimension of training data - unsplit", alltrain.shape)

df_alltrain = pd.DataFrame(alltrain, 
             columns=['ID',
                 'Games Played', 
    'Minutes Played',
    'Points Per Game',
    'Field Goals Made',
    'Field Goals Attempts',
    'Field Goals Percent',
    '3Points Made',
    '3Points Attempts',
    '3Points Percent',
    'Free Throw Made',
    'Free Throw Attempts',
    'Free Throw Percent',
    'Offensive Rebounds',
    'Defensive Rebounds',
    'Rebounds',
    'Assists',
    'Steals',
    'Blocks',
    'Turnovers',
     'TARGET_5Yrs'
                     ])

df_alltrain['TARGET_5Yrs'] = df_alltrain['TARGET_5Yrs'].astype(dtype ='int64')
df_alltrain['ID'] = df_alltrain['ID'].astype(dtype ='str')

df_alltrain.head()
```

## Separate the dependent and independent variables
All features are used

```python
target = df_alltrain.pop('TARGET_5Yrs')
features = df_alltrain[['Games Played', 
    'Minutes Played',
    'Points Per Game',
    'Field Goals Made',
    'Field Goals Attempts',
    'Field Goals Percent',
    '3Points Made',
    '3Points Attempts',
    '3Points Percent',
    'Free Throw Made',
    'Free Throw Attempts',
    'Free Throw Percent',
    'Offensive Rebounds',
    'Defensive Rebounds',
    'Rebounds',
    'Assists',
    'Steals',
    'Blocks',
    'Turnovers']]

# target.info()
# features.info()
```

## Splitting data
Split randomly the dataset with random_state=8 into 2 different sets: training data (80%) and validation data (20%)

```python
X_train, X_val, y_train, y_val = train_test_split (features, 
                                                   target,
                                                   test_size=0.2,
                                                   random_state=8
                                                  )

print("Dimension of features training data", X_train.shape)
print("Dimension of target training data", y_train.shape)
print("Dimension of features validation data", X_val.shape)
print("Dimension of targer validation data", y_val.shape)
```

```python
X_train
```

## Model Development

RFC using the balanced mode


```python
df_y_train = pd.DataFrame(y_train, 
             columns=['TARGET_5Yrs'])
df_y_train['TARGET_5Yrs'].value_counts()
```

```python
# instantiate the model (using the "balanced" mode)
randomforest3_train=RandomForestClassifier(n_estimators=2000,
                                           class_weight = 'balanced',
                                           criterion = 'log_loss',
                                           random_state=7,
                                           max_depth=7
                                          )

# fit the model with data
randomforest3_train.fit(X_train, y_train)
```

```python
# save the model to disk
my_model_filename = '../models/randomforest3.sav'
joblib.dump(randomforest3_train, my_model_filename)
```

## Model Evaluation

```python
# Load the model from the file 
randomforest3_from_joblib = joblib.load('../models/randomforest3.sav') 
```

```python
# Predict on training set
y_pred_ontrain = randomforest3_from_joblib.predict(X_train)

# Accuracy
print( accuracy_score(y_pred_ontrain, y_train) )
```

### Prediction on validation data

```python
y_pred_onval = randomforest3_from_joblib.predict(X_val)
y_pred_onval
```

### Confusion Matrix

```python
cnf_matrix = metrics.confusion_matrix(y_val, y_pred_onval)
cnf_matrix
```

```python
class_names = [0,1] # name  of classes
fig, ax = plt.subplots()
tick_marks = np.arange(len(class_names))
plt.xticks(tick_marks, class_names)
plt.yticks(tick_marks, class_names)

# create heatmap
sns.heatmap(pd.DataFrame(cnf_matrix), annot=True, cmap="YlGnBu" ,fmt='g')
ax.xaxis.set_label_position("top")
plt.tight_layout()
plt.title('Confusion matrix', y=1.1)
plt.ylabel('Actual label')
plt.xlabel('Predicted label');
```

### Accuracy, Precision, and Recall

```python
target_names = ['career years played < 5', 'career years played >= 5']
print(classification_report(y_val, y_pred_onval, target_names=target_names))
```

### ROC

```python
# y_pred_proba = logreg.predict_proba(X_test)[::,1]
y_proba_onval = randomforest3_from_joblib.predict_proba(X_val)[::,1]  # official
fpr, tpr, _ = metrics.roc_curve(y_val,  y_proba_onval)
auc = metrics.roc_auc_score(y_val, y_proba_onval)

plt.plot(fpr,tpr,label="data 1, auc="+str(auc))
plt.legend(loc=4)
plt.show()
```
